```python
#drop rows containing missing values (NaN) and create a new list 
df.dropna(subset=['Item01'], inplace=True)
df_list = df.T.apply(lambda x: x.dropna().tolist()).tolist()
```

```python
#save to new data frame 
df_cleaned = pd.DataFrame(df_list)
```

```python
#OneHot Encoding 
encoder = TransactionEncoder()
onehot = encoder.fit(df_list).transform(df_list)
onehot = pd.DataFrame(onehot, columns=encoder.columns_)
```

```python
#save Oneone-hot encoded boolean to new file
onehot.to_csv('onehot.csv')
```

```python
# Use the Apriori algorithm to generate frequent itemsets
frequent_itemsets = apriori(onehot, use_colnames=True, min_support=0.001, max_len =3)
```

```python
#generate a df of association rules from the frequent itemsets generated by the Apriori algorithm
rules = association_rules(frequent_itemsets)
rules
```

```python
#save columns to new value from rules df
values_1 = (rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
values_1
```

```python
#sorts the rules df by the support column in descending order
rules.sort_values('support', ascending=False)
```

```python
#sorts the rules DataFrame by the confidence column in descending order
rules.sort_values('confidence', ascending=False)
```
