```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
file_path = "/Users/igmark/Desktop/WGU Data Files/D210_churn_clean.csv"
df = pd.read_csv(file_path)
```

```python
df.info
```

```python
df.info(file_path)
```

```python
df.isnull().sum()
```

```python
df.duplicated()
```

```python
# check for outliers and remove
print(df.shape)
df = df[(np.abs(stats.zscore(df.select_dtypes(include=np.number))) < 3).all(axis=1)]
print(df.shape)
```

```python
df.head()
```

```python
#Drop the less meaningful columns

df = df.drop(columns=['Interaction', 'UID', 'Customer_id', 
                      'TimeZone', 'Job','Marital', 'Port_modem', 'Tablet', 'InternetService', 'Phone', 'Multiple',
                      'OnlineSecurity', 'OnlineBackup', 'Area','DeviceProtection', 'StreamingTV', 'StreamingMovies',
                       'Item1', 'Item2', 'Item3', 'Item4', 'Item5', 'Item6', 'Item7', 'Item8'])
```

```python
df = df.rename(columns={'CaseOrder': 'id'})
```

```python
df.info(file_path)
```

```python
df.columns
```

## Saved cleaned data to new file

```python
df.to_csv('D210prepared_churn.csv', index=False)
```
