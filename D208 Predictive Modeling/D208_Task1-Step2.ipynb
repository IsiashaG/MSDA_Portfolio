# Linear Regression

```python
import statsmodels.api as sm

# Set up your independent and dependent variables
X = df[['Z_Score_Children', 'Z_Score_Tenure', 'Z_Score_Yearly_equip_failure', 'Z_Score_MonthlyCharge', 'Z_Score_Age', 
        'Z_Score_Income', 'Z_Score_Email','Z_Score_Contacts', 'Z_Score_Outage_sec_perweek', 
        'Z_Score_Population', 'Z_Score_Phone', 'Z_Score_TechSupport', 'Techie', 'Gender_F', 'Gender_M']]
y = df['Z_Score_Bandwidth_GB_Year']

# Add a constant term to the independent variable matrix
X = sm.add_constant(X)

# Fit an OLS regression model on the data
model = sm.OLS(y, X).fit()

# Print the model summary
print(model.summary())
```

# Reduced Model with VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Define the independent and dependent variables
X = df[['Z_Score_Children', 'Z_Score_Tenure', 'Z_Score_Yearly_equip_failure', 'Z_Score_MonthlyCharge', 'Z_Score_Age', 
        'Z_Score_Income', 'Z_Score_Email','Z_Score_Contacts', 'Z_Score_Outage_sec_perweek', 
        'Z_Score_Population', 'Z_Score_Phone', 'Z_Score_TechSupport', 'Techie', 'Gender_F', 'Gender_M']]
y = df['Z_Score_Bandwidth_GB_Year']

# Calculate VIFs for each independent variable
vifs = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

# Print the VIFs for each variable
for i, col in enumerate(X.columns):
    print(f'{col}: {vifs[i]:.2f}')

# Add a constant term to the independent variables
X = sm.add_constant(X)

# Fit the OLS model
model = sm.OLS(y, X).fit()

# Print the model summary
print(model.summary())
```
